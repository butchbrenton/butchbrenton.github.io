# Opioids

```python
#read in data
df_opioid = pd.read_csv ('OpidsVA.csv') #Note: Filename is different than in printed assignment
# create figure
fig, ax = plt.subplots()
# plot scatterplot
ax.scatter(df_opioid['MedianHouseholdIncome'], df_opioid['FPOO-Rate'])

fig.suptitle('Opioid Overdose Rate vs. Median Houshold Income for Counties in VA')
ax.xaxis.set_label_text('Median Houshold Income')
ax.yaxis.set_label_text('Opioid Overdose Rate')

fig.set_size_inches(7,5)
plt.show()
```

A scatterplot was chosed to compare the number of Opioid Overdoses to Median Household Income.  A scatterplot shows the relationship between two things, and is a good method to identify a correlation which we needed to determine in order to answer the questions. 

Questions: 
1. Do opioid overdoses tend to be assoiciated with less affluent areas - that is, areas where families have lower incomes?
Yes, the opioid overdose rate shows a clear correlation to median household income, whereby counties with a lower median income have a higher rate of opioid overdoses.

2. Some people who start with opioid addictions are reported to transition to heroin use.  What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?
Reviewing the data it is not clear if there is a relationship between opioid overdoses and heroin/fentanyl overdoses.  Although that seems reasonalbe, in order to see a correlation another scatterplot was developed (below),  This graph does not show a correlation between these overdose rates.  


```python
fig, ax = plt.subplots()
ax.scatter(df_opioid['FFHO-Rate'], df_opioid['FPOO-Rate'])

fig.suptitle('Opioid Overdose Rate vs. Heroin Overdose Rate for Counties in VA')
ax.xaxis.set_label_text('Heroin Overdoses')
ax.yaxis.set_label_text('Opioid Overdoses')

fig.set_size_inches(7,5)
plt.show()
```

# Presidents

```python
#read in data
df_pres = pd.read_csv ('presidents.csv') 

#create dictionary of counts
num = df_pres['State']
num = dict(num.value_counts())
#set x and y data
numx = num.keys()
numy = num.values()
#plot bar chart data to make histogram
fig, ax = plt.subplots()
ax.bar(numx,numy)
fig.set_size_inches(7,5)
fig.suptitle('Histogram of US Presidents birth State')
ax.xaxis.set_label_text('Birth State')
plt.xticks(rotation=270)
ax.yaxis.set_label_text('Number of Presidents')
plt.show()

```

A Histogram was chosen to determine in which states the greatest number of US presidents were born.  A histogram is the best way to display frequency of occurence.  

Question:
    
1. Which states are associated with the greatest number of United States presidents in terms of the presidents birthplace?

The US State with the greatest number of presidents born there is Virginia with eight, followed closely by Ohio where seven presidents were born.  Only four states (Virginia, Ohio, New York, and Massachusetts) had more than two presidents born there. 


# US Vehicle Sales

```python
#Read data from file
df_sales = pd.read_csv ('TOTALNSA.csv')
#create graph
fig, ax = plt.subplots()
# plot line graph
ax.plot(df_sales['DATE'], df_sales['TOTALNSA'])
fig.suptitle('United States Vehicle Sales per Month')
ax.xaxis.set_label_text('Date')
ax.yaxis.set_label_text('Number of Vehicles Sold')

fig.set_size_inches(15,5)
plt.show()

```

A Line graph was selected to show the number of vehicles sold each month over time.  A line graph is particularly useful when looking at time series date because trends over time can be quickly seen.

Question:
    
1. How have vehicle sales in the United States varied over time?

The US vehicle sales since 1976 has been cyclic over time, staying within a consistent range.  There are periods of increasing sales are followed by periods of decreasing sales that follow the macroeconomic conditions but the overall trend has been relatively flat.  Along with the macro trends there is also an annual cycle of sales which remains fairly consistent year to year.
